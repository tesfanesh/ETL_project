```python
# Name of the CSV file
file3 ='FastFood_Count & Obesity Rate.csv'
```

```python
# define df
obesityrate_df ='FastFood_Count & Obesity Rate.csv'
```

```python
obesityrate_df = pd.read_csv(file3)
obesityrate_df.head(50)
```

```python
# Show plot
ax1 = obesityrate_df.plot.bar(x="state_name", y=["ff_count"], figsize=(15,4), color='g', align='edge')
ax2 = obesityrate_df.plot.line(x="state_name", y="obesity", color='b', rot=90, ax=ax1, secondary_y=True)
plt.xticks(np.arange(len(obesityrate_df["state_name"]))+0.2, obesityrate_df["state_name"])
plt.xlim(-.5, len(obesityrate_df["state_name"]))
ax1.legend(loc='center', bbox_to_anchor=(.06, 1.1), ncol=1)
ax2.legend(loc='center', bbox_to_anchor=(.94, 1.06), ncol=1)
ax1.set_xlabel("StateName")
ax1.set_ylabel('FastFood Location', fontsize=15)
ax2.set_ylabel('Obesity%', fontsize=15)

plt.title ("FastFood_Count & Obesity Rate Per State")

plt.show()
```
